```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 1000

r_init, theta_init = 0, 0
r_last, theta_last = 2, 2*np.pi
dr, dtheta = 0.05, 0.05

f_edge = lambda theta: 100
f_init = lambda r: 0
f_last = lambda r: 100
```

```python
def polar_laplace(r_init, r_last, dr, theta_init, theta_last, dtheta, f_edge, f_init, f_last, N):
    
    r_space, theta_space = np.arange(r_init, r_last + dr, dr), np.arange(theta_init, theta_last + dtheta, dtheta)
    r_len, theta_len = r_space.size, theta_space.size

    sol = np.ones((r_len, theta_len))

    sol[-1] = f_edge(theta_space)
    sol[:, 0] = f_init(r_space)
    sol[:, -1] = f_last(r_space)
    
    for it in range(N):
        for x in range(1, r_len-1):
            for y in range(1, theta_len-1):
                sol[x, y] = (sol[x-1, y] + sol[x+1, y] + sol[x, y-1] + sol[x, y+1])/4
                
    fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
    ax.contourf(np.linspace(0, 2, theta_len), np.linspace(0, 2, r_len), sol)
                
polar_laplace(r_init, r_last, dr, theta_init, theta_last, dtheta, f_edge, f_init, f_last, N)
            
# fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
# ax.contourf(np.linspace(0, 2, y_len), np.linspace(0, 2, x_len), sol)

plt.show()
```
