# Graph Log-Log Resistance vs Temperature

## State Equation

$$Log(R) = 4.697 - 3.917Log(T)$$

## Code

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Data
R = np.geomspace(200, 30000)
T = np.exp((4.697-np.log(R))/3.917)

#Ploting
plt.loglog(T[::2], R[::2], "or", label='Sample')
plt.loglog(T, R, "-b", label='Function')
plt.legend()
plt.ylabel(r"Resistance $R$ $(\Omega)$")
plt.xlabel(r"Temperature $T$ $(K)$")
plt.title("Resistance of a doped Germanium crystal")
plt.savefig("LogLogRT.pdf")
```
