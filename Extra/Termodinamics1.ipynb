## 1.4 Resistance of a doped Germanium crystal

### State Equation

$$Log(R) = 4.697 - 3.917Log(T)$$

### Code

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Data
R = np.geomspace(200, 30000)
T = np.exp((4.697-np.log(R))/3.917)

#Ploting
plt.loglog(T[::2], R[::2], "or", label='Sample')
plt.loglog(T, R, "-b", label='Function')
plt.legend()
plt.ylabel(r"Resistance $R$ $(\Omega)$")
plt.xlabel(r"Temperature $T$ $(K)$")
plt.title("Resistance of a doped Germanium crystal")
plt.savefig("germanium.pdf")
```

## 1.5 Resistance of a Platinium wire

### Data

- $R = 11000 \Omega$ at the ice point ($0.000^\circ C$)
- $R = 15247 \Omega$ at the steam point ($99.974^\circ C$)
- $R = 28887 \Omega$ at the sulfur point ($444.6^\circ C$)

#### Equation:
$$ R(\theta)=R_0(1+a \theta + b \theta^2)$$

#### Solving for a and b
- $a \approx 3.921 \times 10^{-3}$
- $b \approx -5.934 \times 10^{-7}$

### Code

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Constants
a = 0.003921234183958572
b = -5.933642309051776e-07
R0 = 11000

sample_t = 0, 99.974, 444.6
sample_R = 11000, 15247, 28887

# Data
def R_(t):
    return R0*(1+a*t+b*t**2)

t = np.linspace(0, 660)
R = R_(t)

# Ploting
plt.plot( sample_t, sample_R, "or", label="Sample")
plt.plot(t, R, "-b", label="Function")
plt.legend()
plt.title("Resistance of a Platinium wire")
plt.xlabel(r"Temperature $\theta$ ($^\circ C$)")
plt.ylabel(r"Resistance $R$ ($\Omega$)")
plt.text(270, 20000, r"$ R(\theta)=R_0(1+a \theta + b \theta^2)$", size="large")
plt.savefig("platinium.pdf")
```
