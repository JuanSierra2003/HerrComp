```python
#librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Se definen las constantes de importancia
eps_0, mu_0 = 8.8541878176*10**(-12), 4*np.pi*10**(-7)
mu_p = 0.9999768*mu_0
f_i, f_f = 50*10**12, 300*10**12
d = 10**(-7)
Z_0 = np.sqrt(mu_0/eps_0)
#Se definen las funciones de Transmitancia y Reflectancia, ademas de la permitividad 
def eps_function(f):
    plasmon_frequency = 2.321*10**15
    damping = 5.513*10**12
    return eps_0*(1 - ((plasmon_frequency**2)/(4*np.pi*f**2 + 2j*np.pi*f*damping)))
def Transmittance(f):
    eps_p = eps_function(f)
    Z_p = np.sqrt(mu_p/eps_p)
    theta = d*2j*np.pi*f*np.sqrt(eps_p*mu_p)
    return np.absolute((4/(np.exp(-theta)*(1+(Z_0/Z_p))*(1+(Z_p/Z_0))+ np.exp(theta)*(1+(Z_0/Z_p))*(1-(Z_p/Z_0)))))**2
def Reflectance(f):
    eps_p = eps_function(f)
    Z_p = np.sqrt(mu_p/eps_p)
    theta = d*2j*np.pi*f*np.sqrt(eps_p*mu_p)
    return np.absolute((Z_0**2-Z_p**2)*(np.exp(-theta)-np.exp(theta))/(np.exp(theta)*(Z_0+Z_p)**2 - np.exp(-theta)*(Z_0-Z_p)**2))**2
```

```python
#Valores de la frecuencia asignados a las funciones
F = np.linspace(f_i, f_f, 500)
T = Transmittance(F)
R = Reflectance(F)
```

```python
#Se grafican las figuras
fig = plt.figure(figsize=(10, 8))
plt.plot(F, T, label='Transmitancia')
plt.plot(F, R, label="Reflectancia")
plt.legend()
plt.title("Transmitancia y Reflectancia para una lamina de Plata(Ag)")
plt.xlabel(r"$f$ $(Hz)$")
plt.grid(True)
plt.show()
```
