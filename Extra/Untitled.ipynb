```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
eps_0, mu_0 = 8.8541878176*10**(-12), 4*np.pi*10**(-7)
eps_p, mu_p = 5*10**(-10), 0.99998*mu_0
f_i, f_f = 50*10**12, 300*10**12
d = 10**(-7)

Z_0 = np.sqrt(mu_0/eps_0)
Z_p = np.sqrt(mu_p/eps_p)

def Transmittance(f):
    theta = d*2j*np.pi*f*np.sqrt(eps_p*mu_p)
    return np.absolute(4*Z_0*Z_p/(np.exp(theta)*(Z_0+Z_p)**2 - np.exp(-theta)*(Z_0-Z_p)**2))**2

def Reflectance(f):
    theta = d*2j*np.pi*f*np.sqrt(eps_p*mu_p)
    return np.absolute((Z_0**2-Z_p**2)*(np.exp(-theta)-np.exp(theta))/(np.exp(theta)*(Z_0+Z_p)**2 - np.exp(-theta)*(Z_0-Z_p)**2))**2
```

```python
F  = np.linspace(f_i, f_f, 500)
T = Transmittance(F)
R = Reflectance(F)
```

```python
fig = plt.figure(figsize=(10, 4))

ax1 = fig.add_subplot(1,2,1)
ax1.plot(F, T, label='Transmitancia')
ax1.legend()
ax1.set_ylabel(r"T")
ax1.set_xlabel(r"$f$ $(Hz)$")

ax2 = fig.add_subplot(1,2,2)
ax2.plot(F, R, "r", label='Reflectancia')
ax2.set_ylabel(r"R")
ax2.set_xlabel(r"$f$ $(Hz)$")

ax2.legend()
plt.show()
```
