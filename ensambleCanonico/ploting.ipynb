```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Import data
dataV = pd.read_csv('dataV.txt', names = [ 'v'])[ 'v']
dataVx = pd.read_csv('dataVx.txt', names=['vx'])['vx']
```

```python
#Te√≥rica

v0 = 10
N = 25

#vx
vx = np.linspace(-25,25,100)
var = (v0**2)*N/(2*N-1)
gauss = lambda var,x: (2*np.pi*var)**(-0.5)*np.exp(-x**2/(2*var))

#v
v = np.linspace(0,25,100)
dens = lambda var,x: x*np.exp(-x**2/(2*var))/var
```

```python
#Vx

fig = plt.figure(figsize=(5.4, 4))
fig.suptitle("Histograma de frecuencias componente x de la velocidad")

ax1 = fig.add_subplot(1,1,1)
ax1.hist(dataVx, bins=10, density=True, label=r'$frecuencias$')
ax1.plot(vx, gauss(var,vx), label=r'$\rho(v_x)$')
ax1.legend()
ax1.set_xlabel(r"$v_x$")

plt.show()
```

```python
#V

fig = plt.figure(figsize=(5.4, 4))
fig.suptitle("Histograma de frecuencias magnitud de la velocidad")

ax1 = fig.add_subplot(1,1,1)
ax1.hist(dataV, bins=15, density=True, label=r'$frecuencias$')
ax1.plot(v, dens(var, v), label=r'$\rho(v)$')
ax1.legend()
ax1.set_xlabel(r"$v$")

plt.show()
```
