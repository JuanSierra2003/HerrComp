```python
# use fout.write(f"...") to save the data 
def exp_expansion(x, N, fname):
    """Add docs here"""
    total = 0.0
    fout = open(fname, "w")
    
    for ii in range(0, N+1):
        total+=((-x)**(ii))/factorial(ii)
        fout.write(f"%010d \t %.16f \n"%(ii, total))
        
    fout.close()

def factorial(n):
    """Add docs here"""
    result = 1.0
    for ii in range(1,n+1):
        result *= ii
    return result
 
# Call the function  
exp_expansion(x=1.8766, N=10000, fname="sumdata.txt")
```

```python
# use fout.write(f"...") to save the data 
def exp_expansion_new(x, N, fname):
    """Add docs here"""
    total = 0.0
    a_i = 1.0
    total2 = 0.0
    fout = open(fname, "w")
    for ii in range(0, N+1):
        total+=((-x)**(ii))/factorial(ii)
        total2 += a_i
        a_i*=(-x)/(ii+1)
        fout.write(f"%010d \t %.16f \t %.16f \n"%(ii, total, total2))
    fout.close()

# Call the NEW function  
exp_expansion_new(x=1.8766, N=30, fname="sumdatanew.txt")
```
