```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 1000

x_len = 60
y_len = 50

sol = np. zeros((x_len, y_len))

T_top = 30
T_bottom = 100
T_left = 100
T_right = 20

sol.fill((T_top + T_bottom + T_left + T_right)/4)
sol[0].fill(T_top)
sol[-1].fill(T_bottom)
sol[:, 0].fill(T_left)
sol[:, -1].fill(T_right)


plt.imshow(sol)
plt.colorbar()
plt.show()
```

```python
for it in range(N):
    for x in range(1, x_len-1):
        for y in range(1, y_len-1):
            sol[x, y] = (sol[x-1, y] + sol[x+1, y] + sol[x, y-1] + sol[x, y+1])/4
            
plt.imshow(sol)
plt.colorbar()

fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
ax.contourf(np.linspace(0, 2, y_len), np.linspace(0, 2, x_len), sol)

plt.show()
```
