# Numerical Differentiation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative as spderiv

def forward(f, x, h):
    return (f(x+h)-f(x))/h

def central(f, x, h):
    return (f(x+h/2)-f(x-h/2))/h

def central_richardson(f, x, h):
    return 4*central(f, x, h/2)/3 - central(f, x, h)/3

def compare_derivatives():    
    fig, ax = plt.subplots()
    h = np.power(10.0, np.arange(-18, 0, +1))
    x = np.pi/3
    
    ax.loglog(h, np.abs((forward(np.sin, x, h)-np.cos(x))/np.cos(x)), "b-o", label="Forward")
    ax.loglog(h, np.abs((central(np.sin, x, h)-np.cos(x))/np.cos(x)), "r-s", label="Central")
    ax.loglog(h, np.abs((central_richardson(np.sin, x, h)-np.cos(x))/np.cos(x)), "g-*", label="Central_richardson")
    
    ax.legend()
    plt.xlabel("Particion h")
    plt.ylabel(r"Error $\epsilon$")
    plt.title("Error relativo de diferentes algotitmos de derivada")
    
    plt.savefig("error_derivadas.pdf")
    
compare_derivatives()
```
