```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def pi_MC(N):
    n_circ = 0
    for index in range(N):
        x, y = np.random.rand(2)
        if x**2 + y**2 <= 1: n_circ += 1
    return n_circ*4/N

print(pi_MC(10**6))
```

```python
M, N = 10000, 1000

Data_PIMC = np.zeros(M)
for index in range(M):
    Data_PIMC[index] = pi_MC(N)
```

```python
plt.plot(Data_PIMC)
plt.ylabel(r"PI_MC",size=20)
plt.show()
```

```python
np.mean(Data_PIMC)
```

```python
np.std(Data_PIMC)
```

```python
nbins = int(M**0.5)
lens, base = np.histogram(Data_PIMC, bins=nbins)
```

```python
X = (base[:-1] + base[1:])/2
```

```python
plt.plot(X, lens, "k.")
plt.xlabel(r"$x = \pi _{MC}$", size = 20)
plt.ylabel(r"$f(x)$", size = 20)
plt.show()
```

```python
def pi_MC(N):
    X, Y = np.random.rand(2, N)
    return sum(X**2 + Y**2 <=1)*4/N
```

```python
M, N = 100000, 1000

Data_PIMC = np.zeros(M)
for index in range(M):
    Data_PIMC[index] = pi_MC(N)
```

```python
plt.plot(Data_PIMC)
plt.ylabel(r"PI_MC",size=20)
plt.show()
```

```python
nbins = int(M**0.5)
lens, base = np.histogram(Data_PIMC, bins=nbins)
X = (base[:-1] + base[1:])/2
```

```python
plt.plot(X, lens, "k.")
plt.xlabel(r"$x = \pi _{MC}$", size = 20)
plt.ylabel(r"$f(x)$", size = 20)
plt.show()
```
