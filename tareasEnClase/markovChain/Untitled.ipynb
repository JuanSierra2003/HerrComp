```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_ds = pd.read_csv("Datos Mayo 19 2023/dis_us_d.csv")
```

```python
df_ds.head()
```

```python
plt.plot(df_ds['Open'])
plt.show()
```

```python
prices_dis = np.array(df_ds['Open'])

rel_change = prices_dis[1:]/prices_dis[:-1] -1
```

```python
plt.plot(rel_change)
plt.show()
```

```python
lens, intervals = np.histogram(rel_change, bins=int(rel_change.size**0.5))
base = (intervals[1:]+intervals[:-1])/2

plt.plot(base, lens, 'k.')
```
