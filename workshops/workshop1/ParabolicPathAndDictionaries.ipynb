# Parabolic Path and Dictionaries

El obejetivo es escribir una función que reciba en un diccionario las condiciones iniciales de un lanzamiento parabólico y devuelve la altura máxima alcanzada en dicho lanzamiento. Una vez creada la función se busca crear una gráfica generando datos a partir la misma con una variación del ángulo de inclinación del lanzamiento entre 0 y pi medios y con una velocidad incial igual a 2.34m/s.
A continuación se muestra paso a paso la solución del problema planteado:

- Se importa la libreria numpy y el módulo pyplot de la librería matplotlib para el tratamiento de python numérico y la realización de gráficas respectivamente.

```python
import numpy as np
from matplotlib import pyplot as plt
```

- Se define la constante de la magnitud del campo gravitacional sobre la superficie de la tierra y la velocidad inicial propuesta por el enunciado del problema en unidades del sistema métrico internacional.

```python
g, v0 = 9.8, 2.34
```

- Se define la función max_heigh(dic): que recibe un diccionario con las claves "v0" y "theta0", las cuales representan respectivamente la velocidad inicial y el ángulo de lanzamiento de un movimiento parabólico y retorna la altura máxima alcanzada en dicho movimiento.

```python
def max_heigh(dic):
    return ((dic["v0"]*np.sin(dic["theta0"]))**2)/(2*g)
```

- Se crean dos arreglos de numpy que representan una partición regular en el intervalo [0, pi/2] y los valores de la función max_heigh() en los puntos de dicha partición con una velocidad inicial de 2.34m/s.

```python
X = np.linspace(0, np.pi/2, 100)
Y = max_heigh({"theta0":X, "v0": v0})
```

- Se crea y se guarda la gráfica de la función max_heigh(dic): en el dominio de 0 a pi medios con la velocidad inicial de lanzamiento dada por el problema.

```python
fig = plt.plot(X, Y, label = r"$v_0 = {}m/s$".format(v0))
plt.ylabel(r"$h_{max}$")
plt.xlabel(r"$\theta_0$")
plt.legend()
plt.savefig("max_heigh.pdf")
```
